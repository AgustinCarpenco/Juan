# Análisis y limpieza inicial de datos físicos - 1ra evaluación
Este notebook documenta paso a paso la inspección y limpieza del archivo `1ra evaluación.xlsx` para el análisis de rendimiento físico en futbolistas.

```python
import pandas as pd
import numpy as np
import os
```

```python
def cargar_excel(path):
    """Carga un archivo Excel y retorna un DataFrame. Maneja errores de carga."""
    if not os.path.exists(path):
        print(f"[ERROR] El archivo no existe: {path}")
        return None
    try:
        df = pd.read_excel(path)
        print(f"[INFO] Archivo cargado correctamente: {path}")
        return df
    except Exception as e:
        print(f"[ERROR] Fallo al cargar el archivo: {e}")
        return None

def inspeccion_inicial(df):
    """Muestra información básica del DataFrame de forma visual."""
    if df is None:
        print("[ERROR] DataFrame vacío o no cargado.")
        return
    print(f"Dimensiones: {df.shape}")
    display(df.head())
    display(df.tail())
    display(df.info())
    display(df.describe(include='all'))
    print("\nValores nulos por columna:")
    display(df.isnull().sum())
    print("\nDuplicados:", df.duplicated().sum())
```

```python
ruta = "../data/1ra evaluación.xlsx"  # Ajusta el path si es necesario
df = cargar_excel(ruta)
inspeccion_inicial(df)
```

```python
# Leer el archivo con header en la fila 1 (índice 1), para evitar filas vacías o de títulos
df = pd.read_excel("../data/1ra evaluación.xlsx", header=1)
print(f"Dimensiones iniciales: {df.shape}")

# Eliminar columnas completamente vacías o irrelevantes
df = df.dropna(axis=1, how='all')
if 'Unnamed: 0' in df.columns:
    df = df.drop(columns=['Unnamed: 0'])

df.head()
```

```python
# Renombra las columnas clave si aún no lo hiciste
df = df.rename(columns={
    'Deportista': 'jugador',
    'DNI': 'dni'
})

# Agrega la columna 'categoria' si no existe (puedes cambiar "4ta" por la categoría real)
if 'categoria' not in df.columns:
    df['categoria'] = "4ta"  # O asigna según corresponda

# Muestra los identificadores clave
df[['jugador', 'dni', 'categoria']].head()
```

```python
import numpy as np

def obtener_riesgo(row):
    if row.get('Alto Riesgo', 0) == row[['Bajo Riesgo', 'Moderado Riesgo', 'Alto Riesgo']].max():
        return 'Alto'
    elif row.get('Moderado Riesgo', 0) == row[['Bajo Riesgo', 'Moderado Riesgo', 'Alto Riesgo']].max():
        return 'Moderado'
    elif row.get('Bajo Riesgo', 0) == row[['Bajo Riesgo', 'Moderado Riesgo', 'Alto Riesgo']].max():
        return 'Bajo'
    else:
        return np.nan

# Crea la columna Riesgo
df['Riesgo'] = df.apply(obtener_riesgo, axis=1)

# Visualiza el resultado
df[['jugador', 'Bajo Riesgo', 'Moderado Riesgo', 'Alto Riesgo', 'Riesgo']].head(10)
print("Conteo de cada categoría de riesgo:")
print(df['Riesgo'].value_counts(dropna=False))
```

```python
# Lista de valores a excluir (puedes agregar más si detectas otros)
excluir = ['MEDIA', 'SD', 'ALTO RIESGO', 'MODERADO RIESGO', 'BAJO RIESGO', np.nan, '']

# Filtra solo jugadores reales
df = df[~df['jugador'].isin(excluir)]
df = df[df['jugador'].notnull()]  # Elimina filas sin nombre de jugador

# Resetea el índice para orden limpio
df = df.reset_index(drop=True)

# Verifica el resultado
print("Dimensiones después de limpiar filas resumen:", df.shape)
df[['jugador', 'dni', 'categoria', 'Riesgo']].head(10)
print("Conteo de riesgo tras limpieza:")
print(df['Riesgo'].value_counts(dropna=False))
```

```python
import pandas as pd
df = pd.read_csv("../data/1ra_evaluacion_limpia.csv")
```

```python
import matplotlib.pyplot as plt

variables = ['AKE_Der', 'AKE_Izq', 'Cuad_70_Der', 'Cuad_70_Izq', 'Isq_30_Der', 'Isq_30_Izq', 'RSI', 'TS_Der', 'TS_Izq']
for var in variables:
    if var in df.columns:
        plt.figure(figsize=(6,3))
        df[var].hist(bins=15)
        plt.title(f"Histograma de {var}")
        plt.xlabel(var)
        plt.ylabel("Frecuencia")
        plt.show()
```

```python
import seaborn as sns

pares = [('AKE_Der', 'AKE_Izq'), ('Cuad_70_Der', 'Cuad_70_Izq'), ('TS_Der', 'TS_Izq')]
for der, izq in pares:
    if der in df.columns and izq in df.columns:
        plt.figure(figsize=(6,3))
        sns.boxplot(data=df[[der, izq]])
        plt.title(f"Boxplot: {der} vs {izq}")
        plt.show()
```

```python
import seaborn as sns

plt.figure(figsize=(4,3))
sns.countplot(data=df, x='Riesgo', order=['Bajo', 'Moderado', 'Alto'])
plt.title("Distribución de Riesgo")
plt.show()
```

```python
from scipy.stats import zscore

# Ejemplo con Cuad_70_Der
if 'Cuad_70_Der' in df.columns:
    df['Z_Cuad_70_Der'] = zscore(df['Cuad_70_Der'].dropna())
    plt.figure(figsize=(6,3))
    sns.histplot(df['Z_Cuad_70_Der'].dropna(), bins=15, kde=True)
    plt.title("Z-Score Cuad_70_Der")
    plt.xlabel("Z-Score")
    plt.show()
```

```python
if 'TS_LSI' in df.columns:
    plt.figure(figsize=(6,3))
    sns.histplot(df['TS_LSI'].dropna(), bins=15, kde=True)
    plt.axvspan(90, 110, color='green', alpha=0.2, label='Normalidad (90-110%)')
    plt.title("Distribución LSI Triple Salto")
    plt.xlabel("LSI (%)")
    plt.legend()
    plt.show()
```
